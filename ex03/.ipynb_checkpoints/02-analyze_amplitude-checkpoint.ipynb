```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14
%run plot.py

import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten #, Reshape
from keras.layers import Conv1D, MaxPooling1D, AveragePooling1D
```

## Load data (A=500)

```python
# A identifies the amplitude considered in the notebook
# The possibilities are: 300, 500
A = 500

str0 = f'ts_L60_Z12_A{A}_DX50_bias5_N10000.dat'
fnamex = 'DATA/x_' + str0
fnamey = 'DATA/y_' + str0

x = np.loadtxt(fnamex, delimiter=' ', dtype=float)
N, L = len(x), len(x[0])

Show_data(x, 6, 'original data')
print('data: ', N)

# build multi-class vector labels
categ = np.loadtxt(fnamey, dtype=int)
n_class = 3
y = np.zeros((N,n_class))
for i in range(N):
    y[i,categ[i]] = 1
print(y[:6])
```

### Rescale data

```python
# remove average value of each sample from its values
xm = x.mean(axis=1)
for i in range(N):
    x[i] = x[i] - xm[i]

# rescale all the samples with the dataset variance 
xv = x.var()
x = x/np.sqrt(xv)

Show_data(x, 6, 'rescaled data')
```

### Split training, validation and test data

```python
# number of training, validation and test data
perc_train = 0.5
perc_valid = 0.1
N_train = int(perc_train*N)
N_val = int(perc_valid*N)

x_train, y_train = x[0:N_train], y[0:N_train]
x_val, y_val = x[N_train:N_train+N_val], y[N_train:N_train+N_val]
x_test1, y_test1 = x[N_train+N_val:], y[N_train+N_val:]

print('training:', len(x_train),'\nvalidation:', len(x_val), '\ntest:', len(x_test1))

# Keras wants an additional dimension with a 1 at the end
x_train = x_train.reshape(x_train.shape[0], L, 1)
x_val =  x_val.reshape(x_val.shape[0], L, 1)
input_shape = (L, 1)
```

### Define the model with Keras

```python
# reproducibility
np.random.seed(12345)
import tensorflow.random as tf_r
tf_r.set_seed(12345)

# initializer
ini = tf.keras.initializers.RandomNormal(mean=0.0, stddev=0.05, seed=None)
# regularizers
reg = tf.keras.regularizers.l2(0.2)

model1 = Sequential()
# convolutional layer
model1.add(Conv1D(filters=5, kernel_size=11,
                kernel_initializer=ini, kernel_regularizer=reg,
                activation='relu', input_shape=(L,1)))
# pooling layer                     
model1.add(AveragePooling1D(5))
# convolutional layer
model1.add(Conv1D(filters=5, kernel_size=7,
                activation='relu'))
# fully connected neural network
model1.add(Flatten())
model1.add(Dense(12, activation='relu'))
model1.add(Dropout(0.2))        
model1.add(Dense(n_class, activation='softmax'))

# compile the model: categorical_crossentropy, 3 output nodes
model1.compile(loss=keras.losses.categorical_crossentropy,
                optimizer='Adam', metrics=['accuracy'])
```

```python
# clone model1 into model2
# they are identical but trained on different datasets
model2 = keras.models.clone_model(model1)
model2.compile(loss=keras.losses.categorical_crossentropy,
              optimizer='Adam',metrics=['accuracy'])
```

### Define functions to show history

```python
def Show_history(fit):
    fig, AX = plt.subplots(1,2,figsize=(12,5))

    # plot accuracy
    ax = AX[0]
    ax.plot(fit.history['accuracy'], 'b', label='training')
    ax.plot(fit.history['val_accuracy'], 'r--', label='validation')
    ax.plot((0,len(fit.history['accuracy'])), (1/3,1/3), ':', c='gray', label='random choice')
    ax.set_xlabel('epoch')
    ax.set_ylabel('Accuracy')
    ax.set_ylim([0, 1])
    ax.legend()

    # plot loss
    ax = AX[1]
    ax.plot(fit.history['loss'], 'b', label='training')
    ax.plot(fit.history['val_loss'], 'r--', label='validation')
    ax.set_xlabel('epoch')
    ax.set_ylabel('Loss')
    ax.set_ylim([0, 1.05*np.max(fit.history['loss'])])
    ax.legend()

    plt.show()
```

### Training

```python
BATCH_SIZE = 256
EPOCHS = 128
fit1 = model1.fit(x_train,y_train,batch_size=BATCH_SIZE,
                epochs=EPOCHS,
                validation_data=(x_val, y_val),
                verbose=2, shuffle=True) 
```

## Learning curves and confusion matrices on test set

```python
print('----- Model 1 (A=500) -----')
Show_history(fit1)
print('----- Model 2 (A=300) -----')
Show_history(fit2)
```

```python
import pandas as pd
from scipy import stats
import seaborn 
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn import preprocessing


def show_confusion_matrix(test, pred):

    LABELS = ['absent','positive','negative']
    cmap='GnBu'

    cm = metrics.confusion_matrix(test, pred)
    
    plt.figure(figsize=(6, 5))
    seaborn.heatmap(cm,
                xticklabels=LABELS,
                yticklabels=LABELS,
                annot=True,
                fmt='d',
                linecolor='white',
                linewidths=1,
                cmap=cmap)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    return cm

# true labels
max_y_test1 = np.argmax(y_test1, axis=1)
max_y_test2 = np.argmax(y_test2, axis=1)

# predicted labels: take the class with the highest probability

print('----- Model 1 (A=500) -----')
y_pred1 = model1.predict(x_test1)
max_y_pred1 = np.argmax(y_pred1, axis=1)
cm1 = show_confusion_matrix(max_y_test1, max_y_pred1)
print('test accuracy:', np.sum(cm1[0,0] + cm1[1,1] + cm1[2,2])/len(max_y_pred1), '\n')

print('----- Model 1 (A=300) -----')
y_pred2 = model2.predict(x_test2)
max_y_pred2 = np.argmax(y_pred2, axis=1)
cm2 = show_confusion_matrix(max_y_test2, max_y_pred2)
print('test accuracy:', np.sum(cm2[0,0] + cm2[1,1] + cm2[2,2])/len(max_y_pred2), '\n')
```
