### Function for the random step
Each step is obtained by taking the int() of a random value drawn from a Gaussian distribution with constant mean $bias$ and standard deviation $DX$.

```python
# random seed for reproducibility
np.random.seed(12345)

# function for the random step, using lambda construction
# int() for cleaner look and for mimiking a detector with finite resolution
jump = lambda drift, stdev: int(np.random.normal(drift,stdev))

# test the random step function
print([jump(5,50) for i in range(10)])
```

### Function for the added pattern
It adds to the second and third samples of each triplet of data a sinusoidal pattern, positive or negative, of half period duration controlled by $Z$, with amplitude given by $A$.

```python
# function for the pattern
def pattern(x,Z,A):
    return int(A*np.sin((np.pi*x)/Z))

# plot patterns
fig, ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
    sign = 1 - 2*i
    x = np.arange(1,12)
    y = np.array([sign*pattern(j,12,500) for j in x])
    ax[i].plot(x, y)
    ax[i].set_xlabel('time')
    ax[i].set_ylabel('amplitude')
    ax[i].set_title('pattern ' + str(i+1))

plt.show()
```

### Generate data

```python
# random seed for reproducibility
np.random.seed(12345)

# number of steps
Z = 12
# amplitude, the possibilities are: 300, 500
A = 500

# number of samples
N = 10000
# size of each sample of the time-series
L = 60

# step parameters: introduce small positive bias
bias = 5
DX = 50

x = np.zeros((N,L))
y = np.zeros(N)

for i in range(N):

    # connect two adjacent samples
    if i > 0:
        x[i,0] = x[i-1,-1] + jump(bias,DX)
    
    # build the sample
    for j in range(1,L):
        x[i,j] = x[i,j-1] + jump(bias,DX)

    # assign the label to the sample    
    y[i] = i%3 
    #y[i] = random.randint(0,2)

    # add the pattern
    if y[i] > 0:
        j0 = np.random.randint(0,L-1-Z)
        sign = 3 - 2*y[i]
        for j in range(Z):
            x[i,j0+j] += sign*pattern(j,Z,A)
    
Show_data(x, 6, 'original data')

# print some samples          
for i in range(3):
    print(x[i], y[i])
```
