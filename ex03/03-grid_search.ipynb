```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14
%run plot.py

import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten #, Reshape
from keras.layers import Conv1D, MaxPooling1D, AveragePooling1D
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.regularizers import l1, l2, l1_l2
from sklearn.model_selection import KFold
import pandas as pd
```

### Load data

```python
# A identifies the amplitude considered in the notebook
# The possibilities are: 300, 500
A = 500

str0 = f'ts_L60_Z12_A{A}_DX50_bias5_N10000.dat'
fnamex = 'DATA/x_' + str0
fnamey = 'DATA/y_' + str0

x = np.loadtxt(fnamex, delimiter=' ', dtype=float)
N, L = len(x), len(x[0])

Show_data(x, 6, 'original data')
print('data: ', N)

# build multi-class vector labels
categ = np.loadtxt(fnamey, dtype=int)
n_class = 3
y = np.zeros((N,n_class))
for i in range(N):
    y[i,categ[i]] = 1
print(y[:6])
```

### Rescale data

```python
# remove average value of each sample from its values
xm = x.mean(axis=1)
for i in range(N):
    x[i] = x[i] - xm[i]

# rescale all the samples with the dataset variance 
xv = x.var()
x = x/np.sqrt(xv)

Show_data(x, 6, 'rescaled data')
```

### Split training, validation and test data

```python
# number of training, validation and test data
perc_train = 0.5
perc_valid = 0.1
N_train = int(perc_train*N)
N_val = int(perc_valid*N)

x_train, y_train = x[0:N_train], y[0:N_train]
x_val, y_val = x[N_train:N_train+N_val], y[N_train:N_train+N_val]
x_test, y_test = x[N_train+N_val:], y[N_train+N_val:]

print('training:', len(x_train),'\nvalidation:', len(x_val), '\ntest:', len(x_test))

# Keras wants an additional dimension with a 1 at the end
x_train = x_train.reshape(x_train.shape[0], L, 1)
x_val =  x_val.reshape(x_val.shape[0], L, 1)
```

### Define the model with Keras

```python
def create_model(reg='l2', lamb=0.1):
    
    # reproducibility
    np.random.seed(12345)
    import tensorflow.random as tf_r
    tf_r.set_seed(12345)

    # initializer
    ini = tf.keras.initializers.RandomNormal(mean=0.0, stddev=0.05, seed=None)

    # regularizers
    if reg == 'l1':
        reg = l1(lamb)
    if reg == 'l2':
        reg = l2(lamb)

    model = Sequential()
    # convolutional layer
    model.add(Conv1D(filters=5, kernel_size=11,
                    kernel_initializer=ini, kernel_regularizer=reg,
                    activation='relu', input_shape=(L,1)))
    # pooling layer                     
    model.add(AveragePooling1D(5))
    # convolutional layer
    model.add(Conv1D(filters=5, kernel_size=7,
                    activation='relu'))
    # fully connected neural network
    model.add(Flatten())
    model.add(Dense(12, activation='relu'))
    model.add(Dropout(0.2))        
    model.add(Dense(n_class, activation='softmax'))

    # compile the model: categorical_crossentropy, 3 output nodes
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='Adam', metrics=['accuracy'])
    
    return model
```

### Define the grid search parameters

```python
# KerasClassifier wraps Keras models for use in scikit-learn 
model = KerasClassifier(build_fn=create_model)

param_grid = {
    'reg' : ['l1','l2'],
    'lamb' : [0, 0.5, 0.1, 0.01, 0.001,],
    'batch_size' : [256],
    'epochs' : [128]
}
```

### Training

```python
cv = KFold(n_splits=3, random_state=12345, shuffle=True)
final_model = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=cv, verbose=1)
final_model.fit(x_train, y_train)
```

### Results

```python
means = final_model.cv_results_['mean_test_score']
stds = final_model.cv_results_['std_test_score']
params = final_model.cv_results_['params']

df = pd.DataFrame(params)
df.insert(0, 'score', np.round(means, 5))
df.insert(1, 'std', np.round(stds, 4))
df['lamb']=df['lamb']
df.sort_values(by=['score'], ascending=False)
```

### DISCUSSION

From the results of the performed GridSearch it is possible to see that the best performances are reached with the Ridge kernel regularizer with coefficient $\lambda_{2}=0.1$ . This model is the one chosen as the final one. For higher and lower values of $\lambda_{2}$ the average cross-validated  scores (namely the accuracy) are worse. Concerning the Lasso kernel regularizer, the maximum score in the grid is achieved for $\lambda_{1}=0.01$ .

### Define functions to show history and weights

```python
def Show_history(fit):
    fig, AX = plt.subplots(1,2,figsize=(12,5))

    # plot accuracy
    ax = AX[0]
    ax.plot(fit.history['accuracy'], 'b', label='training')
    ax.plot(fit.history['val_accuracy'], 'r--', label='validation')
    ax.plot((0,len(fit.history['accuracy'])), (1/3,1/3), ':', c='gray', label='random choice')
    ax.set_xlabel('epoch')
    ax.set_ylabel('Accuracy')
    ax.set_ylim([0, 1])
    ax.legend()

    # plot loss
    ax = AX[1]
    ax.plot(fit.history['loss'], 'b', label='training')
    ax.plot(fit.history['val_loss'], 'r--', label='validation')
    ax.set_xlabel('epoch')
    ax.set_ylabel('Loss')
    ax.set_ylim([0, 1.05*np.max(fit.history['loss'])])
    ax.legend()

    plt.show()

def Show_weights(model, l=0):
    # define colors and markers
    c = ['r','y','c','b','m','g','k']
    m = ['o','s','D','<','>','p','P']
    ms = 10
    
    # get weights
    w = model.layers[l].get_weights()[0]
    wT = w.T
    M = len(wT)
    # get bias
    b = model.layers[l].get_weights()[1]
    
    fig, AX = plt.subplots(1,2,figsize=(12,4.4))

    # plot weights
    ax = AX[0]
    ax.axhline(0, c='k')
    ax.plot((0,))
    for i in range(M):
        ax.plot(wT[i,0], '-', c=c[i], marker=m[i], label=str(i), markersize=ms)
    ax.set_title('Filters of layer ' + str(l+1))
    ax.set_xlabel('index')

    # plot bias
    ax = AX[1]
    ax.axhline(0, c='k')
    for i in range(M):
        ax.plot((i), (b[i]), c=c[i], marker=m[i], label='filter ' + str(i), markersize=ms)
    ax.set_title('Bias of layer ' + str(l+1))
    ax.set_xlabel('filter nr')
    ax.set_xticks(np.arange(5))
    ax.legend()

    plt.show()
```

### Train a bad model

```python
# bad model
model = create_model(reg='l1', lamb=0.1)
print(model.summary())

# train bad model
fit = model.fit(x_train, y_train, batch_size=256,
                epochs=128,validation_data=(x_val, y_val),
                verbose=2, shuffle=True)

Show_history(fit)
Show_weights(model)
```

### Train the best model

```python
# best model
model = create_model(reg='l2', lamb=0.1)
print(model.summary())

# train best model
fit = model.fit(x_train, y_train, batch_size=256,
                epochs=128,validation_data=(x_val, y_val),
                verbose=2, shuffle=True)

Show_history(fit)
Show_weights(model)
```

### DISCUSSION
Above are displayed the trained weights of the filters of the first convolutional layer for a model with unsatisfactory performance (with the Lasso regularizer with $\lambda_{1}=0.1$) and for the final chosen model (with the Ridge regularizer with $\lambda_{2}=0.1$). It can be seen that the last one not only achieves higher accuracy but also provides a meaningful visualization of the weights. In fact the shapes of the weights of the three filters are similar to the sinusoidal pattern considered as the signal to be found. On the other hand, for the first model, the visualization of the weights explains why the accuracy is lower: only one filter has learned the patterns to be searched in the samples.

### Confusion matrix on test set

```python
from scipy import stats
import seaborn 
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn import preprocessing


def show_confusion_matrix(test, pred):

    LABELS = ['absent','positive','negative']
    cmap='GnBu'

    cm = metrics.confusion_matrix(test, pred)
    
    plt.figure(figsize=(6, 5))
    seaborn.heatmap(cm,
                xticklabels=LABELS,
                yticklabels=LABELS,
                annot=True,
                fmt='d',
                linecolor='white',
                linewidths=1,
                cmap=cmap)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    return cm


# true labels
max_y_test = np.argmax(y_test, axis=1)

# predicted labels: take the class with the highest probability
y_pred = model.predict(x_test)
max_y_pred = np.argmax(y_pred, axis=1)

cm = show_confusion_matrix(max_y_test, max_y_pred)
print('test accuracy:', np.sum(cm[0,0] + cm[1,1] + cm[2,2])/len(max_y_pred))
```
