```python
import numpy as np
from numpy import cos, sin, pi, random
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
plt.rcParams['font.size'] = 14
mycmap = ListedColormap(["blue","red","gold" ])

!mkdir DATA
```

```python
np.random.seed(12345)

# number of data samples
N=1000

# size of each sample: we generate the data for L=3, L=4, L=6
L=3
# We introduced noisy inputs in the additional dimensions. 
# This noise is extracted by a standard normal distrubution multiplied by 10 for L=4, by 1 or by 10 for L=6
noise_amp_3=np.array([10,10,10])
noise_amp_4=np.array([10,10,10,10])
n_amp_6 = 1 # or 10
noise_amp_6=np.array([10,10,10,n_amp_6,n_amp_6,n_amp_6])

#  % of points in cluster 0 and 1
p0,p1=35,45

# radii for 3d construction
R0=800
R1=400

x,y,r,noise=np.zeros((N,L)),np.zeros(N),np.random.randint(0,100,N),np.random.randn(N,L)

for n in range(N):
    if r[n]<p0:
        y[n]=0
        q = random.rand()*2*pi
        x[n,:3] = (R0*cos(q), R0*sin(q), -(R0/10)*cos(6*q))
    elif r[n]<p1+p0:
        y[n]=1
        q = random.rand()*2*pi
        x[n,:3] = (-(R1/5)*sin(4*q), R1*cos(q), R1*sin(q))
        x[n,:3] = x[n,:3] + R0*np.array((0,random.randint(3) - 1, 0))
    else:
        y[n]=2
        q = random.rand()*1000.
        x[n,:3] = ((random.randint(2)-0.5)*2*q, 1.333*q-R1, 1.333*R1-0.333*q)
        
x = x + noise*noise_amp_3
y = y.astype(int)


# plot

fig = plt.figure(figsize=(15, 5))
for i in range(3):
    ax = fig.add_subplot(1,3,i+1,projection='3d')
    ax.scatter(x[:,0], x[:,1],x[:,2], s=5, c=y, depthshade=True, cmap=mycmap)
    ax.view_init(15+15*i, 45*(i-1))
plt.draw()

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection='3d')
ax.scatter(x[:,0], x[:,1],x[:,2], s=30, c=y, depthshade=True, cmap=mycmap)
ax.view_init(20, 40)
plt.draw()

if L==6:
    str0 = f'3d_L{L}_N_amp{n_amp_6}.dat'
else:
    str0 = f'3d_L{L}.dat'

print(str0)

fname = 'DATA/x_' + str0
np.savetxt(fname,x,delimiter='\t')
fname = 'DATA/y_' + str0
np.savetxt(fname,y,fmt='%d')
```
