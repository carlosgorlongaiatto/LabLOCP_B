```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.ticker import NullFormatter
plt.rcParams['font.size'] = 14

# manifold.TSNE
from sklearn import manifold  #, datasets
from time import time
```

## Data and labels 
In this example we know the labels, hence we can test the performances of t-SNE and DBSCAN

```python
fname,fnamey="x_3d_L3.dat","y_3d_L3.dat"
data=np.loadtxt("DATA/"+fname, delimiter='\t')
y=np.loadtxt("DATA/"+fnamey,dtype=int)
N=len(data)
D=len(data[0])
print("N =",N,"  D =",D)
x=data
```

## Discussion

### t-SNE 

The t-SNE method allows to visualize datasets with many feutures and therefore many dimensions in a space with a smaller dimension. The algorithm first projects all the points in a space with the desired dimension (n-components) through the instruction given in the function 'init'. Then it proceeds to the rearrangement of the points on this new space with the aim of improving the interpretability of the dataset while trying to minimize the loss of information in the transition from the space with the highest dimension to the one with the lowest dimension.

### Random initialization

Using init='random' each dataset point in the multidimensional feuture space is randomly assigned to a point in the n-component dimensional space with distribution proportional to $N(0,1)$.


### Perplexity

Another important tunable parameter of t-SNE is "perplexity" which tells how to balance the focus between local and global aspects of the data. This parameter is, in some ways, an assumption about the number of nearby neighbors each point has.
As it can be seen from the graph below with a low perplexity $(\sim5)$ t-SNE shows many small groups well separated from each other; with an intermediate perplexity $(\sim20)$ there are a smaller number of well separated groups; with a high perplexity $(\sim200)$ there is only one group of points all connected to each other. 
Groups are intended here as a visual pattern because t-SNE is not a clustering algorithm but only a visualization tool.
An important information is that the perplexity cannot be greater than the number of points in the dataset, otherwise the algorithm will fail. 


```python
np.random.seed(12345)
n_samples = N
n_components = 2 
perplexities = [5, 10, 20, 50, 120, 200]
Np = len(perplexities)
mycmap = ListedColormap(["blue","red","gold" ])

(fig, subplots) = plt.subplots(1,Np , figsize=(3*Np, 2.7))

X_all = [None]
for i, perplexity in enumerate(perplexities):
    
    tsne = manifold.TSNE(n_components=n_components, init='random',
                         random_state=0, perplexity=perplexity)
    Y = tsne.fit_transform(data)

    ax = subplots[i]
    ax.set_title("Perplexity=%d" % perplexity)
    ax.scatter(Y[:, 0], Y[:, 1], c=y,cmap=mycmap)
    ax.tick_params(left=False,bottom=False)
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    ax.axis('tight')
    X_all.append(Y)
    
plt.show()
```

### PCA initialization

Using init='pca' the points in the multidimensional feutures space are projected into the n-components dimensional space made up of the n-components principal components of the PCA performed on the original dataset.

### Perplexity

As in random initialization, the clarity of the patterns decreases as the perplexity increases. In this case the impact of perplexity is noticeable but less evident than in the previous case. This is probably due to the fact that for our dataset the PCA decomposition is sufficient to distinguish different patterns.

```python
from sklearn.decomposition import PCA

model = PCA(n_components = 2)
y_pca = model.fit_transform(x)

print(y.shape)
plt.scatter(y_pca[:,0], y_pca[:,1])
plt.show()
```
