```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, b=20, w=10, q=0.1):
    return b*(1 - np.exp((-0.5*w*(x[0]**2 + x[1]**2)))) + 0.5*q*(-x[0]**3 + x[1])**2

def grad_f(x, b=20, w=10, q=0.1):
    grad_x = b*np.exp(-0.5*w*(x[0]**2 + x[1]**2))*w*x[0] - 3*q*(x[0]**2)*(-x[0]**3 + x[1])
    grad_y = b*np.exp(-0.5*w*(x[0]**2 + x[1]**2))*w*x[1] + q*(-x[0]**3 + x[1])
    return np.array([grad_x, grad_y])
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y, sparse=True)
z = f((X, Y))

h = ax.contour(x, y, z, levels=20, cmap='coolwarm')
fig.colorbar(h)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

<b>Gradient Descent</b>

$\theta_{t+1}= \theta_t - \eta_t \nabla_\theta E(\theta_t)$

```python
def GD(grad, init, eta, epochs):
    theta = init
    theta_list = np.zeros([epochs+1,2])
    theta_list[0,:] = theta
    v = 0
    for i in range(0, epochs):
        v = eta*grad(theta)
        theta = theta - v
        theta_list[i+1,:] = theta
    return theta_list
```

```python
np.random.seed(1234)
init = 6*np.random.random(size=2) - 3
eta = 0.001
epochs = 200
theta_list = GD(grad_f, init, eta, epochs)

print('theta_init:', init)
print('theta_best:', theta_list[-1,:])
print('f(theta_best):', f(theta_list[-1,:]))

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(np.arange(0, theta_list.shape[0]), [f(x) for x in theta_list])
ax.set_yscale('log')
ax.set_xlabel('epoch')
ax.set_ylabel('cost function')
ax.set_title('GD')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y, sparse=True)
z = f((X, Y))

h = ax.contour(x, y, z, levels=20, cmap='coolwarm')
fig.colorbar(h)
ax.plot(theta_list[:,0], theta_list[:,1], '--', c='b')
ax.plot(theta_list[-1,0], theta_list[-1,1], 'o', c='r', ms=6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('GD')
plt.show()
```

<b>Gradient Descent with Momentum</b>

$v_{t}=\gamma v_{t-1}+\eta_{t}\nabla_\theta E(\theta_t),\\
\theta_{t+1}= \theta_t -v_{t}$

```python
def GDM(grad, init, eta, epochs, gamma):
    theta = init
    theta_list = np.zeros([epochs+1,2])
    theta_list[0,:] = theta
    v = 0
    for i in range(0, epochs):
        v = gamma*v + eta*grad(theta)
        theta = theta - v
        theta_list[i+1,:] = theta
    return theta_list
```

```python
np.random.seed(1234)
init = 6*np.random.random(size=2) - 3
eta = 0.001
epochs = 200
gamma = 0.9
theta_list = GDM(grad_f, init, eta, epochs, gamma)

print('theta_init:', init)
print('theta_best:', theta_list[-1,:])
print('f(theta_best):', f(theta_list[-1,:]))

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(np.arange(0, theta_list.shape[0]), [f(x) for x in theta_list])
ax.set_yscale('log')
ax.set_xlabel('epoch')
ax.set_ylabel('cost function')
ax.set_title('GDM')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y, sparse=True)
z = f((X, Y))

h = ax.contour(x, y, z, levels=20, cmap='coolwarm')
fig.colorbar(h)
ax.plot(theta_list[:,0], theta_list[:,1], '--', c='b')
ax.plot(theta_list[-1,0], theta_list[-1,1], 'o', c='r', ms=6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('GDM')
plt.show()
```
