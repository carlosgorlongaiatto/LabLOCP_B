```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# fix the seed for reproducibility
np.random.seed(12345)
```

```python
# number of data
N = [4000, 1000, 8000]
N_label = ['normal', 'reduced', 'increased']
# sample size
L = 2

!mkdir DATA
```

### Define the labelling function

```python
# labelling function
def f(x, c=1):
    r = 0
    if c == 1:
        if x[0] > -20 and x[1] > -40 and x[0] + x[1] < 40:
            r = 1
    if c == 2:
        if (np.sign(x.sum())*np.sign(x[0]))*np.cos(np.linalg.norm(x)/(2*np.pi)) > 0:
            r =  1
    return r
```

### Generate data

```python
# box size
B = 100
# function type
c = 1

fig, ax = plt.subplots(1, 3, figsize=(16,5))

for i in range(3):

    # generate 2D points in a grid of size (-B/2,B/2)*(-B/2,B/2)
    x, y = (np.random.random((N[i],2)) - 0.5)*B, np.zeros(N[i])

    for j in range(N[i]):
        # label data according to the labelling function
        y[j] = f(x[j], c)

    np.savetxt('./DATA/data-' + N_label[i] + '.dat', x)
    np.savetxt('./DATA/labels-' + N_label[i] + '.dat', y, fmt='%d')

    ax[i].scatter(x[:,0], x[:,1], c=y, cmap='viridis')
    ax[i].set_title(N_label[i] + ', N = ' + str(len(x)),fontsize=18)
    ax[i].set_xlabel('$x_1$', fontsize=18)
    ax[i].set_ylabel('$x_2$', fontsize=18, labelpad=-15)
    ax[i].tick_params(labelsize=14, pad=5)
    
plt.savefig('dataset.pdf')
plt.show()
```
