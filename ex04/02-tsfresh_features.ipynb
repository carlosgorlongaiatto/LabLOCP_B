```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14

from xgboost import XGBClassifier, plot_tree
```

### Load data

```python
str0 = 'ts_L60_Z12_A500_DX50_bias5_N20000.dat'
fnamex = 'DATA/x_' + str0
fnamey = 'DATA/y_' + str0

x = np.loadtxt(fnamex, delimiter=" ",dtype=float)
N, L = len(x), len(x[0])

# function to plot samples
def Show_data(x, N, s='data'):
    # size of each sample of the timeseries
    L = 60
    # plot the first N samples
    for i in range(N):
        plt.plot(np.arange(i*L,(i+1)*L),x[i])   
    plt.xlabel('time')
    plt.title(s)
    plt.show()

Show_data(x, 6, 'original data')
print('data:', N)

# note: here it does not need to be converted to the 3-bit version, a label remains y[i]=0,1,2
y = np.loadtxt(fnamey, dtype=int)
n_class = 3
print(y[:6])
```

### Load tsfresh features

```python
# load dataframe created in notebook '01-cnn_vs_xgboost'
x_features = pd.read_csv('x_features.csv')

# look at x_features
x_features
```

### Split training and test set

```python
perc_train = 0.8
N_train = int(perc_train*N)
x_train = x_features[:N_train]
y_train = y[:N_train]
x_test = x_features[N_train:]
y_test = y[N_train:]
N_test = len(x_test)

print('training:', N_train, '\ntest:', N_test)
```

### Rescale data

```python
# remove train set average value of each feature
xm = np.mean(x_train,axis=0)
x_train = x_train - xm
x_test = x_test - xm

# rescale with the train set standard deviation
xv = np.var(x_train,axis=0)
x_train = x_train/np.sqrt(xv)
x_test = x_test/np.sqrt(xv)
```

```python
# remove columns with NaN or inf        
x_train.replace([np.inf,-np.inf], np.nan, inplace=True)
x_train = x_train.dropna(axis='columns')

x_test.replace([np.inf,-np.inf], np.nan, inplace=True)
x_test = x_test.dropna(axis='columns')

# look at train set
x_train
```

### xgboost

```python
# reproducibility
np.random.seed(12345)

# define parameters for xgboost
params = {'max_depth':6,'min_child_weight':1,
        'learning_rate':0.3,'use_label_encoder':False}

# build model with given params
model = XGBClassifier(**params)

# fit
model.fit(x_train.values,y_train)
```

```python
import graphviz 

# plot tree
fig, AX = plt.subplots(3,1,figsize=(30, 10))
for i in range(3):
    ax=AX[i]
    plot_tree(model, num_trees=i, ax=ax)
    
fig.savefig('xgboost-trees.png', dpi=300, pad_inches=0.02)
```

### Accuracy and confusion matrix

```python
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix

LABELS = ['absent','positive','negative']
cmap = 'GnBu'

def show_confusion_matrix(validations, predictions):

    matrix = confusion_matrix(validations, predictions)
    plt.figure(figsize=(6, 5))
    sns.heatmap(matrix,
                xticklabels=LABELS,
                yticklabels=LABELS,
                annot=True,
                fmt='d',
                linecolor='white',
                linewidths=1,
                cmap=cmap)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# predict labels on test set
y_pred = model.predict(x_test)

# compute accuracy
acc_test = accuracy_score(y_test,y_pred) 
print('accuracy:', acc_test)

show_confusion_matrix(y_test,y_pred)
```

### Feature importances

See an overview of extracted features:
https://tsfresh.readthedocs.io/en/latest/text/list_of_features.html

```python
# get feature names
feature_names = x_train.columns.values

# empty dict
feat_imp = {}

# get feature importance from model
for i,f in enumerate(feature_names):
    feat_imp[f] = model.feature_importances_[i]

# sort features depending on their importances
feat_imp = dict(sorted(feat_imp.items(), reverse=True, key=lambda item: item[1]))
```

```python
minval = 0.005
print('Feature importance sum = ' + str(np.sum(model.feature_importances_)))
print('Showing feature with importance > ' + str(minval))
for i,k in enumerate(feat_imp.keys()):
    if feat_imp[k] > minval: print(i+1,k,feat_imp[k])
```

### Visualize features

```python
# get feature names given importance order
features = list(feat_imp.keys())

# build dataframe for sns pairplot
df = pd.DataFrame({'1':x_features[features[0]],
                   '2':x_features[features[1]],
                   '3':x_features[features[2]],
                   '4':x_features[features[3]],
                   '5':x_features[features[4]],
                   'Class':y})
# pairplot with seaborn
pal = sns.blend_palette(['blue','red','gold'], n_class)
sns.pairplot(df, hue='Class', plot_kws=dict(alpha=0.8), palette=pal);
```

### Explain feature "change_quantiles"
https://tsfresh.readthedocs.io/en/latest/_modules/tsfresh/feature_extraction/feature_calculators.html#change_quantiles

```python
# build the change_quantiles function by hand
def change_quantiles(ts, ql=0.0, qh=0.5, isabs=False, f_agg='var', isprint=False):

    # compute differences/changes
    div = np.diff(ts)
    if isabs:
        div = np.abs(div)

    # quantile-based discretization
    bin_cat = pd.qcut(ts, [ql, qh], labels=False)
    bin_cat_0 = bin_cat == 0

    # count differences/changes that start and end inside the corridor
    ind = (bin_cat_0 & np.roll(bin_cat_0, 1))[1:]

    # get position of differences inside the corridor
    ind_inside_corridor = np.where(ind == 1)

    # result
    aggregator = getattr(np, f_agg)
    result = aggregator(div[ind_inside_corridor])
    print('feature value by hand:', result)

    # plot
    plt.plot(ts, '-o')
    y0 = np.min(ts)
    y_range = np.max(ts) - np.min(ts)
    yl, yh = y0+ql*y_range, y0+qh*y_range
    plt.axhline(y=yl, c='r')
    plt.axhline(y=yh, c='r')
    plt.fill_between(np.arange(len(ts)), yl, yh, color='r', alpha=0.1)
    plt.xlabel('time')
    plt.show()

    # print all the details
    if isprint:
        print('time-series:\n', ts)
        print('differences:\n', div)
        print('values of the time-series inside the corridor:\n', bin_cat, '\n', bin_cat_0)
        print('differences inside the corridor:\n', ind)
        print('positions:\n', ind_inside_corridor)
        print('differences inside the corridor:\n', div[ind_inside_corridor])
```

```python
name = 'value__change_quantiles__f_agg_"var"__isabs_False__qh_0.6__ql_0.0'
print('feature: ' + name + '\n')

# inspect the first three time-series
for i in range(3):
    print('label:', i)
    print('feature value:', x_features[name][i])
    change_quantiles(x[i], 0.0, 0.6, False, 'var')
```

Four out of first five features are obtained through the "change_quantile" function. We find out that when quantile boundaries are low, the feature is able to separate the third class, while high value boundaries can detect the second class. Since these features are extracted by computing the differences of contiguous values of the time-series, it is possible to understand that the rescaling was not necessary.

### Explain feature "agg_linear_trend"

https://tsfresh.readthedocs.io/en/latest/_modules/tsfresh/feature_extraction/feature_calculators.html#agg_linear_trend

The function "agg_linear_trend" returns an attribute of the least-squares linear regression for values of the time series that have been aggregated over chunks with respect to the sequence from 0 up to the number of chunks minus one. In our case, the most important feature extracted via this function has 5 chunks, the aggregator is the variance, and the attribute is the error of the slope of the linear regression. The result shows that this feature is able to separate the first class from the other ones.
